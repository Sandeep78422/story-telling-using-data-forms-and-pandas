3. Get the visualization of your specific player data using year wise plot.
Create a story on the players performances over the span of years he has
played IPL.
Your specific player is any player of your choice between the serial
numbers (your Roll_No*3 -3) and (your Roll_No*3). For example, if
your Roll No is 20, you may select your player anyone placed between
serial number 58 to 60. 

```python
import matplotlib.pyplot as plt
```

```python

selected_player = 'Ricky Bhui' 
player_data = df[df['Player_Name'] == selected_player]


plt.figure(figsize=(10, 6))
plt.plot(player_data['Year'], player_data['Runs_Scored'], marker='o', label='Runs Scored')
plt.plot(player_data['Year'], player_data['Catches_Taken'], marker='s', label='Catches Taken')
plt.plot(player_data['Year'], player_data['Wickets_Taken'], marker='^', label='Wickets Taken')

plt.title(f"Year-Wise Performance of {selected_player}")
plt.xlabel("Year")
plt.ylabel("Performance Metrics")
plt.legend()
plt.grid()
plt.show()
```

In the above graph, we can observe that Ricky bhui year wise performance is better at the year 2019 as compared to the other years' performance, but here we can see that his performance increased rapidly and started decreasing gradually at the year 2019.

4. Create a story on comparison of the performance of your player with the
other two players at serial numbers between (your Roll_No*3 -3) and
(your Roll_No*3). Select proper plots for visualization. 

```python

comparison_players = ['Ricky Bhui', 'Kumar Kushagra', 'Tristan Stubbs']
comparison_data = df[df['Player_Name'].isin(comparison_players)]

plt.figure(figsize=(12, 8))
for player in comparison_players:
    player_df = comparison_data[comparison_data['Player_Name'] == player]
    plt.plot(player_df['Year'], player_df['Runs_Scored'], marker='o', label=f"{player} - Runs")

plt.title("Year-Wise Comparison of Runs Scored")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.xlim(2017,2025)
plt.grid()
plt.show()
```

In the above graph, we can see that Ricky Bhui performance is average. but in the case of Stubbs initially, it increased and for the next year it decreased.

5. Create a story around year wise changes in the overall IPL
performances. For that, get the records of year wise total run scored,
total catches and total wickets taken, including all the player records.
Comment on the year-wise variations in them and support your
comments using proper visualizations. 

```python

yearly_data = df.groupby('Year').sum()

plt.figure(figsize=(12, 6))
plt.plot(yearly_data.index, yearly_data['Runs_Scored'], label='Total Runs', marker='o')
plt.plot(yearly_data.index, yearly_data['Catches_Taken'], label='Total Catches', marker='s')
plt.plot(yearly_data.index, yearly_data['Wickets_Taken'], label='Total Wickets', marker='^')

plt.title("Year-Wise Overall IPL Performance Trends")
plt.xlabel("Year")
plt.ylabel("Total Metrics",fontsize=14)
plt.legend()
plt.grid()
plt.xlim(2014,2024)
plt.show()

print("Observations on year-wise IPL performance trends:")
print("1. Runs scored have shown consistent growth/decline over certain years.")
print("2. Catches and wickets show fluctuations, possibly due to changes in match dynamics.")
```
